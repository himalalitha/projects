### DATA PREPROCESSING

```python

un_data.dropna(inplace=True)  
```

```python
un_data.columns = un_data.columns.str.lower()
un_data.isnull().sum()
```

```python
un_data
```

```python
cleaned_data = un_data.dropna()
```

```python
cleaned_data.columns = cleaned_data.columns.str.strip()
```

```python
cleaned_data['date'] = cleaned_data['date'].str.strip()
```

```python
cleaned_data['date'] = pd.to_datetime(cleaned_data['date'], format='%d-%m-%Y')
```

```python

unemployment_stats = cleaned_data['estimated unemployment rate (%)'].describe()
```

```python
print(cleaned_data.info())
print(unemployment_stats)
```

### DATA VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=cleaned_data, x='date', y='estimated unemployment rate (%)', hue='region')
plt.title('Time Series of Unemployment Rate in India by Region')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=cleaned_data, x='region', y='estimated unemployment rate (%)')
plt.title('Distribution of Unemployment Rates by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['estimated unemployment rate (%)'], bins=20, kde=True)
plt.title('Histogram of Unemployment Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=cleaned_data, x='estimated labour participation rate (%)', y='estimated unemployment rate (%)', hue='region')
plt.title('Unemployment Rate vs Labour Participation Rate')
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
plt.figure(figsize=(14, 7))
stacked_data = cleaned_data.pivot_table(values='estimated unemployment rate (%)', index='date', columns='region', aggfunc='mean')
stacked_data.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Stacked Bar Graph of Unemployment Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=cleaned_data, x='region', y='estimated unemployment rate (%)', hue='area')
plt.title('Unemployment Rate by Region and Area')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
region_unemployment = cleaned_data.groupby('region')['estimated unemployment rate (%)'].mean()
```

```python
plt.figure(figsize=(6, 6))
region_unemployment.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Average Unemployment Rate by Region')
plt.ylabel('') 
plt.show()
```

```python
area_unemployment = cleaned_data.groupby('area')['estimated unemployment rate (%)'].mean()
```

```python
plt.figure(figsize=(4, 4))
area_unemployment.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])
plt.title('Average Unemployment Rate by Area (Rural vs. Urban)')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
numerical_data = cleaned_data[['estimated unemployment rate (%)', 'estimated employed', 'estimated labour participation rate (%)']]
```

```python
correlation_matrix = numerical_data.corr()
```

```python
plt.figure(figsize=(5, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap of Correlations Between Numerical Variables')
plt.show()
```
